## Matplotlib example workflow

```python
# 0. import matlplotlib and get it ready fo plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(4, 4))

# 3. plot data
ax.plot(x, y)

# 4. customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. save and show (save the whole figure)
fig.savefig("images/samle-plot.png")
```

## Two options for subplots

```python
# Option number 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option number 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Object oriented methods of matplotlib

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# pyplot method
# c stands for colour, it will be coloured on the basis of target column
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
# object oriented methods
fig, ax1 = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax1);
# ax1.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line to represent the average
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

# Customize the ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol");

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a mean line
ax0.axhline(over_50["chol"].mean(), linestyle="--");

#Add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"]);

# Customize the ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target");

# Add a mean line
ax1.axhline(over_50["thalach"].mean(), linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data, cmap -> colur scheme
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter");

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line to represent the average
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter");

# Customize the ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol");

ax0.set_xlim([50,80]);

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a mean line
ax0.axhline(over_50["chol"].mean(), linestyle="--");

#Add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter");

# Customize the ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");

ax1.set_ylim([80, 200])

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target");

# Add a mean line
ax1.axhline(over_50["thalach"].mean(), linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

# Save and Share the plot
fig.savefig("images/heart-disease-analysis-plot.png");
```
